```python
from pathlib import Path
from ase import io

samples_dir = Path("samples")  # 필요 시 절대경로로 교체

for traj_path in samples_dir.glob("*.traj"):
    # 모든 프레임 읽기 (단일 프레임이면 길이 1 리스트)
    frames = io.read(traj_path, index=":")
    base = traj_path.with_suffix("")  # 확장자 제거한 경로

    for i, atoms in enumerate(frames):
        # 프레임이 하나면 원래 이름.cif, 여러 개면 번호 suffix
        cif_path = base.with_suffix(".cif") if len(frames) == 1 else base.with_name(f"{base.name}_{i}").with_suffix(".cif")
        io.write(cif_path, atoms)
        print(f"saved {cif_path}")
```
